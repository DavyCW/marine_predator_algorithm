```python
import pandas as pd
import pickling_inputs
from sklearn.utils import resample

x_train = pd.read_pickle('data pickles/x_train_mc_ecg')
y_train = pd.read_pickle('data pickles/y_train_mc_ecg')
y_train_pd=pd.DataFrame(y_train)
y_train_pd.rename(columns={y_train_pd.columns[0]: "y"}, inplace=True)
train=pd.concat([x_train, y_train_pd], axis=1)
classes = len(pd.unique(train["y"]))
train_classes = []
train_duped_classes = []
train_max = 0
for i in range(classes):
    train_classes.append(train[train['y'] == i])
    if train_max < train_classes[i].shape[0]:
        train_max = train_classes[i].shape[0]
for i in range(classes):
    if train_classes[i].shape[0] == train_max:
        train_duped_classes.append(train_classes[i])
    else:
        train_duped_classes.append(resample(train_classes[i], random_state=0, n_samples=train_max, replace=True))
train_duped = pd.concat(train_duped_classes)
y_duped_train = train_duped['y']
x_duped_train = train_duped.drop(columns=['y'])

pickling_inputs.pickling((x_duped_train, y_duped_train), 
                         ("x_duped_train_mc_ecg", "y_duped_train_mc_ecg"))
```
