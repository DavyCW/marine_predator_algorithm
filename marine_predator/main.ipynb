```python
#from func_plot import Func_Plot
import func_plot
import mpa
import get_function_details
import pandas as pd

# Set parameters
search_agents_no = 25
max_iteration = 24
```

```python
functions = ['F24', 'F25', 'F26']
data = []

temp_data = []
temp_data.append("smote15_train_mctobin_ecg")
temp_data.append(pd.read_pickle('data pickles/x_smote15_train_mctobin_ecg').values)
temp_data.append(pd.read_pickle('data pickles/y_smote15_train_mctobin_ecg').values)
temp_data.append(pd.read_pickle('data pickles/x_test_mc_ecg').values)
temp_data.append(pd.read_pickle('data pickles/y_test_mctobin_ecg').values)
data.append(temp_data)

temp_data = []
temp_data.append("smote7_train_mc_ecg")
temp_data.append(pd.read_pickle('data pickles/x_smote7_train_mc_ecg').values)
temp_data.append(pd.read_pickle('data pickles/y_smote7_train_mc_ecg').values)
temp_data.append(pd.read_pickle('data pickles/x_test_mc_ecg').values)
temp_data.append(pd.read_pickle('data pickles/y_test_mc_ecg').values)
data.append(temp_data)

temp_data = []
temp_data.append("train_bin_ecg")
temp_data.append(pd.read_pickle('data pickles/x_train_bin_ecg').values)
temp_data.append(pd.read_pickle('data pickles/y_train_bin_ecg').values)
temp_data.append(pd.read_pickle('data pickles/x_test_bin_ecg').values)
temp_data.append(pd.read_pickle('data pickles/y_test_bin_ecg').values)
data.append(temp_data)
```

```python
results = []
for f in functions:
    lb, ub, dim, fobj = get_function_details.Get_Function_Details(f)
    print(f)
    for d in data:
        print('\t' + d[0])
        results.append((f, d, mpa.MPA(search_agents_no, max_iteration, lb, ub, dim, fobj, d[1], d[2], d[3], d[4])))
```

```python
# Function topology plot
for result in results:
    func_plot.Func_Plot(result[0], result[2][2], result[2][3], result[1][1], result[1][2], result[1][3], result[1][4])
```

```python
# Display results
for result in results:
    print(f'For function {result[0]} on the {result[1][0]} dataset, we have the following results:')
    print(f'The best solution obtained by MPA is : {result[2][1]}')
    print(f'The best optimal value of the objective function found by MPA is : {result[2][0]}')
    print('--------------------------------------')
```
